```python
# mapping quarter:
quarter_description = {
    1:'1st Quarter',
    2:'2nd Quarter',
    3:'3rd Quarter',
    4:'4th Quarter'
}
```

```python
# extract data:
import pandas as pd

url='https://github.com/CSSEGISandData/COVID-19/raw/master/csse_covid_19_data/csse_covid_19_time_series/'
df_global_confirmed = pd.read_csv(url+'time_series_covid19_confirmed_global.csv', index_col=[                                                  'Province/State', 'Country/Region', 'Lat', 'Long'                 ]).stack().to_frame()
df_global_recovered = pd.read_csv(url+'time_series_covid19_recovered_global.csv', index_col=[                                                  'Province/State', 'Country/Region', 'Lat', 'Long'                 ]).stack().to_frame()
df_global_deaths    = pd.read_csv(url+'time_series_covid19_deaths_global.csv'   , index_col=[                                                  'Province/State', 'Country/Region', 'Lat', 'Long'                 ]).stack().to_frame()
df_us_confirmed     = pd.read_csv(url+'time_series_covid19_confirmed_US.csv'    , index_col=['UID', 'iso2', 'iso3', 'code3', 'FIPS', 'Admin2', 'Province_State', 'Country_Region', 'Lat', 'Long_', 'Combined_Key']).stack().to_frame()
df_us_deaths        = pd.read_csv(url+'time_series_covid19_deaths_US.csv'       , index_col=['UID', 'iso2', 'iso3', 'code3', 'FIPS', 'Admin2', 'Province_State', 'Country_Region', 'Lat', 'Long_', 'Combined_Key']).stack().to_frame()
```

```python
# transform and clean 'global':
df_global = pd.concat([df_global_confirmed, df_global_recovered, df_global_deaths], axis=1).fillna(0).reset_index()
df_global.drop(columns=['Lat', 'Long'], inplace=True) # we create latitude and longitude with geopy
df_global.columns = ['State', 'Country', 'Date', 'Confirmed_Total', 'Recovered_Total', 'Deaths_Total']
df_global['Country'].replace({'Timor-Leste'  :'East Timor',
                              "Cote d'Ivoire":'Ivory Coast',
                              'Burma'        :'Myanmar',
                              'Korea, South' :'South Korea',
                              'Taiwan*'      :'Taiwan',
                              'Holy See'     :'Vatican City'}, inplace=True)
df_global['State'].fillna(df_global['Country'], inplace=True)
df_global = df_global.astype({'Date':'datetime64'})
df_global
```

```python
# transform and clean 'US':
df_us = pd.concat([df_us_confirmed, df_us_deaths], axis=1).fillna(0).reset_index()
df_us.drop(columns=['UID', 'iso2', 'iso3', 'code3', 'FIPS', 'Admin2', 'Lat', 'Long_', 'Combined_Key'], inplace=True)
df_us.columns = ['State', 'Country', 'Date', 'Confirmed_Total', 'Deaths_Total']
df_us['Country'].replace({'US':'United States of America'}, inplace=True)
df_us = df_us[df_us['Date']!='Population'] # we don't use population in this model
df_us = df_us.astype({'Date':'datetime64'})
df_us
```

```python
# merge 'global' and 'US' into 'staging':
df_staging = pd.concat([df_global, df_us]).fillna(0).reset_index(drop=True)

# map continent:
df_staging['Continent'] = df_staging['Country'].map(country_continent)

# sort and aggregate:
df_staging = (df_staging.sort_values(by=['Continent', 'Country', 'State', 'Date'])
                        .groupby(['Continent', 'Country', 'State', 'Date'])
                        .agg(sum)).reset_index()

# add unique country ID:
df_staging['State'].mask(df_staging['State']==df_staging['Country'], '', inplace=True)
df_staging['Country-ID'] = df_staging['Continent']+' '+df_staging['Country']+' '+df_staging['State']
df_staging['Country-ID'] = (df_staging['Country-ID'].str.strip()
                                                    .str.lower()
                                                    .str.replace( ' ', '-', regex=False)
                                                    .str.replace( ',', '-', regex=False)
                                                    .str.replace( '(', '-', regex=False)
                                                    .str.replace( ')', '-', regex=False)
                                                    .str.replace('--', '-', regex=False)
                                                    .str.strip('-'))

# add latitude and longitude:
from time import sleep
from geopy.exc import GeocoderTimedOut
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="covid19-star-schema")

def geolocate(request, attempt=1, max_attempts=5):
    sleep(0.3) # delay to reduce timeouts
    try:
        return geolocator.geocode(request)
    except GeocoderTimedOut:
        if attempt <= max_attempts:
            return geolocate(request, attempt=attempt+1)
        raise

df_geo = pd.DataFrame({'Country-ID':df_staging['Country-ID'],
                       'Country':df_staging['Country'],
                       'State':df_staging['State'],
                       'Latitude':0.0,
                       'Longitude':0.0}).drop_duplicates().reset_index(drop=True)

for index, row in df_geo.iterrows():
    location = geolocate(row['Country']+' '+row['State'])
    if (location is not None) and (row['Country-ID'] not in ['ships-diamond-princess', 'ships-ms-zaandam']):
        df_geo.at[index,'Latitude'] = location.latitude
        df_geo.at[index,'Longitude'] = location.longitude

df_geo.drop(columns=['Country', 'State'], inplace=True)
df_geo.to_csv('data/MAP_Geo.csv')

df_staging['Latitude'] = df_staging['Country-ID'].map(dict(zip(df_geo['Country-ID'], df_geo['Latitude'])))
df_staging['Longitude'] = df_staging['Country-ID'].map(dict(zip(df_geo['Country-ID'], df_geo['Longitude'])))

# restructure:
df_staging = df_staging.reindex(columns=['Country-ID', 'Continent', 'Country', 'State', 'Latitude', 'Longitude', 'Date', 'Confirmed_Total', 'Recovered_Total', 'Deaths_Total'])
```

```python
# add measurement 'absolute values per day':
df_staging['Confirmed_AbsPerDay'] = df_staging.groupby(['Continent', 'Country', 'State'], as_index=False)[['Confirmed_Total']].diff().fillna(0).astype('int64')
df_staging['Recovered_AbsPerDay'] = df_staging.groupby(['Continent', 'Country', 'State'], as_index=False)[['Recovered_Total']].diff().fillna(0).astype('int64')
df_staging[   'Deaths_AbsPerDay'] = df_staging.groupby(['Continent', 'Country', 'State'], as_index=False)[[   'Deaths_Total']].diff().fillna(0).astype('int64')

# add measurement 'average values per day':
df_staging['Confirmed_AvgPerDay'] = df_staging.groupby(['Continent', 'Country', 'State'], as_index=False)[['Confirmed_AbsPerDay']].rolling(7).mean().fillna(0).round(decimals=2).reset_index(drop=True) # we use a 7 day rolling average
df_staging['Recovered_AvgPerDay'] = df_staging.groupby(['Continent', 'Country', 'State'], as_index=False)[['Recovered_AbsPerDay']].rolling(7).mean().fillna(0).round(decimals=2).reset_index(drop=True)
df_staging[   'Deaths_AvgPerDay'] = df_staging.groupby(['Continent', 'Country', 'State'], as_index=False)[[   'Deaths_AbsPerDay']].rolling(7).mean().fillna(0).round(decimals=2).reset_index(drop=True)

# add measurement 'days since 1000 confirmed cases':
df_staging['Confirmed_DaysSince1000'] = df_staging.loc[df_staging['Confirmed_Total']>=1000].groupby(['Continent', 'Country', 'State'], as_index=False).cumcount()+1
df_staging['Recovered_DaysSince1000'] = df_staging.loc[df_staging['Recovered_Total']>=1000].groupby(['Continent', 'Country', 'State'], as_index=False).cumcount()+1
df_staging[   'Deaths_DaysSince1000'] = df_staging.loc[df_staging[   'Deaths_Total']>=1000].groupby(['Continent', 'Country', 'State'], as_index=False).cumcount()+1
df_staging.fillna(0, inplace=True)
```

```python
# load staging table:
df_staging = df_staging.astype({'Confirmed_Total':'int64', 'Recovered_Total':'int64', 'Deaths_Total':'int64', 'Confirmed_DaysSince1000':'int64', 'Recovered_DaysSince1000':'int64', 'Deaths_DaysSince1000':'int64'})
df_staging.to_csv('data/Staging.csv')
df_staging
```

```python
# load dimension 'Country':
df_Country = pd.DataFrame({'Country-ID':df_staging['Country-ID'],
                           'State':df_staging['State'],
                           'Country':df_staging['Country'],
                           'Continent':df_staging['Continent'],
                           'Total':'Total',
                           'Latitude':df_staging['Latitude'],
                           'Longitude':df_staging['Longitude']}).drop_duplicates().reset_index(drop=True)
df_Country.loc[df_Country['Country']=='US', 'Total'] = '' # 'US' not part of 'Total' hierarchy
df_Country.to_csv('data/DIM_Country.csv')
df_Country
```

```python
# load dimension 'Time':
df_Time = pd.DataFrame({'Day':df_staging['Date'],
                        'Month':df_staging['Date'].dt.month.map(month_description),
                        'Quarter':df_staging['Date'].dt.quarter.map(quarter_description),
                        'Year':df_staging['Date'].dt.year}).drop_duplicates().reset_index(drop=True)
df_Time.to_csv('data/DIM_Time.csv')
df_Time
```

```python
# load dimension 'DaysSince1000':
df_DaysSince1000 = pd.DataFrame({'DaysSince1000':range(max(df_staging['Confirmed_DaysSince1000'].unique().tolist()
                                                          +df_staging['Recovered_DaysSince1000'].unique().tolist()
                                                          +df_staging[   'Deaths_DaysSince1000'].unique().tolist())),
                                 'Total':'Total'}).drop_duplicates().reset_index(drop=True)
df_DaysSince1000.to_csv('data/DIM_DaysSince1000.csv')
df_DaysSince1000
```

```python
# load dimension 'Cases':
df_Cases = pd.DataFrame({'Cases':['Confirmed', 'Recovered', 'Deaths']})
df_Cases.to_csv('data/DIM_Cases.csv')
df_Cases
```

```python
# load dimension 'Datatype':
df_Datatype = pd.DataFrame({'Datatype':['AbsPerDay', 'AvgPerDay', 'Total']})
df_Datatype.to_csv('data/DIM_Datatype.csv')
df_Datatype
```

```python
# load fact table:
df_Fact = df_staging
df_Fact.set_index(['Country-ID', 'Continent', 'Country', 'State', 'Latitude', 'Longitude', 'Date', 'Confirmed_DaysSince1000', 'Recovered_DaysSince1000', 'Deaths_DaysSince1000'], inplace=True)
df_Fact = df_Fact.stack().to_frame().reset_index()
df_Fact.columns = ['Country-ID', 'Continent', 'Country', 'State', 'Latitude', 'Longitude', 'Time', 'Confirmed_DaysSince1000', 'Recovered_DaysSince1000', 'Deaths_DaysSince1000', 'Cases_Datatype', 'Value']
df_Fact[['Cases','Datatype']] = df_Fact['Cases_Datatype'].str.split('_', expand=True)
df_Fact['DaysSince1000'] = 0
df_Fact.loc[df_Fact['Cases']=='Confirmed', 'DaysSince1000'] = df_Fact['Confirmed_DaysSince1000']
df_Fact.loc[df_Fact['Cases']=='Recovered', 'DaysSince1000'] = df_Fact['Recovered_DaysSince1000']
df_Fact.loc[df_Fact['Cases']==   'Deaths', 'DaysSince1000'] = df_Fact[   'Deaths_DaysSince1000']
df_Fact.drop(columns=['Continent', 'Country', 'State', 'Latitude', 'Longitude', 'Confirmed_DaysSince1000', 'Recovered_DaysSince1000', 'Deaths_DaysSince1000', 'Cases_Datatype'], inplace=True)
df_Fact = df_Fact.reindex(columns=['Country-ID', 'Time', 'DaysSince1000', 'Cases', 'Datatype', 'Value'])
df_Fact = df_Fact[df_Fact['Value']!=0].reset_index(drop=True)
df_Fact.to_csv('data/FACT_Covid19.csv')
df_Fact
```
