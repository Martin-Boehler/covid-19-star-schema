```python
import os
import pymysql
from sqlalchemy import create_engine

# read environment variables:
mysql_host     = os.getenv('MYSQL_HOST',     0)
mysql_port     = os.getenv('MYSQL_PORT',     0)
mysql_database = os.getenv('MYSQL_DATABASE', 0)
mysql_user     = os.getenv('MYSQL_USER',     0)
mysql_password = os.getenv('MYSQL_PASSWORD', 0)

if mysql_host == 0 or mysql_port == 0 or mysql_database == 0 or mysql_user == 0 or mysql_password == 0:
    print('os: error reading environment variables!')
    raise SystemExit

# create sqlalchemy engine
engine = create_engine('mysql+pymysql://{user}:{password}@{host}:{port}/{database}'.format(user=mysql_user, password=mysql_password, host=mysql_host, port=mysql_port, database=mysql_database))
```

```python
import pandas as pd

# load csv and bulk write:
def load(table):
    df = pd.read_csv('/home/jovyan/work/data/' + table + '.csv', index_col=0)
    df.to_sql(name=table, con=engine, if_exists='replace', index=False, )

load('Staging')
load('MAP_Geo')
load('DIM_Cases')
load('DIM_Country')
load('DIM_Datatype')
load('DIM_DaysSince1000')
load('DIM_Time')
load('FACT_Covid19')
```
